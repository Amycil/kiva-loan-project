##LOANS AWARDED BY KIVA BASED ON COUNTRY, GENDER AND ACTIVITY

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib.ticker as mtick
```

##LOAD AND EXAMINE THE DATA.

Every row of the dataset represents a loan that Kiva awarded to a particular project. The  `Loan_amount` column shows the amount awarded in US dollars, the  `activity` column shows the particular project,  `the country` columnn represents the country the project is located and the  `gender` column refers to the primary person to which the loan was awarded or the person that applied for the loan.


```python
kiva = pd.read_csv(r"C:\Users\Surface Pro 3\Documents\kiva loan project\seabornProjectCumulative\kiva_data.csv")
kiva.head()
```

Creating a bar plot using Seaborn to visualize the average size of Kiva loans given to projects, by country.

Using `sns.barplot()` with the following arguments:

- `data` set to `kiva`
- `x` set to `country`
- `y` set to `loan_amount`

```python
f, ax = plt.subplots(figsize=(15, 10))
sns.barplot(data=kiva, x='country', y='loan_amount')

fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
plt.show()
```

**Nesting gender against the loan_amount and country columns**

```python

f, ax = plt.subplots(figsize=(15, 10))

sns.barplot(data=kiva, x="country", y="loan_amount", hue='gender')
sns.set_style('darkgrid')
sns.color_palette('pastel')

#To assign the '$' sign to the amount
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
```

#### Reflection Questions

## BOX PLOT BY ACTIVITY

Instead of visualizing the loan amount by *country*, using  `sns.boxplot()` to plot the loan amount by *activity.

```python
plt.figure(figsize=(16, 10))

sns.boxplot(data=kiva, x='activity', y='loan_amount')
sns.set_style('darkgrid')
sns.color_palette('colorblind')
plt.title('Loan amount By Projects')
plt.show()
```

## Violin Plots

Using `sns.violionplot` to visualize loan_amount by activity.

```python
plt.figure(figsize=(16, 10))

sns.violinplot(data=kiva, x="activity", y="loan_amount")
plt.show()
```

### Creating a violin plot that visualizes the distribution of loan amount by country.
Previously, I created a violin plot and plotted the data by _activity_. This time, a violin plot that plots the data by _country_.


```python
plt.figure(figsize=(16, 10))
sns.violinplot(data=kiva, x='country', y='loan_amount')
plt.show()
```

## Split Violin Plots By Gender

Using the `hue` and `split` parameters with `sns.violinplot()` to visualize the distribution of loan amount by country, split by gender. 



```python
sns.violinplot(data=kiva, x='country', y='loan_amount', hue='gender', split=True)
sns.set_palette("Spectral")
plt.figure(figsize=(16, 12))
plt.show()
```

#### Reflection Questions

What does this visualization reveal about the distribution of loan amounts within countries by gender?

- Males are awarded more loans than females within the countries.
